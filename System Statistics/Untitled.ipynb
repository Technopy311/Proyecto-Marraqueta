```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress
```

```python
df = pd.read_csv('/home/technopy/Documents/Proyecto-Marraqueta/ProyectoMarraqueta/increasing_timings.csv')
```

```python
time_std = df["Time(s)"].std()
```

```python
slope, intercept, r_value, p_value, std_err = linregress(df["Users#"], df["Time(s)"])
```

```python
plt.figure(figsize=(12, 5))
text_content = f"y={slope}x + {intercept}"

plt.plot(df["Users#"], intercept + slope*df["Users#"], 'r', label='Trend Line')
plt.scatter(df["Users#"], df["Time(s)"], color='blue', s=1.5)

plt.xlabel("Iteration #")
plt.ylabel("Execution Time (s)")

plt.text(180, 0.0827, text_content, fontsize = 8)

plt.legend()
plt.show()
```
